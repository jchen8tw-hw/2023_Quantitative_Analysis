```python
import pandas as pd
import plotly.express as px
import numpy as np
pd.options.plotting.backend = "plotly"
```

# Read in Data

```python
df = pd.read_csv("./data/rating.csv",index_col=0,dtype={"稅後純益(百萬)": str,"稅後EPS(元)":str})
df["稅後純益(百萬)"].replace("- -",pd.NA,inplace=True)
df["稅後EPS(元)"].replace("- -",pd.NA,inplace=True)
df["稅後純益(百萬)"] = pd.to_numeric(df["稅後純益(百萬)"],).astype(pd.Int64Dtype())
df["稅後EPS(元)"] = pd.to_numeric(df["稅後EPS(元)"])
df["股票代號"] = df["股票代號"].astype(str)
df["日期"] = pd.to_datetime(df["日期"])
```

```python
group_by_stock = df.groupby("股票代號")
```

```python
group_by_stock.get_group("2330")["券商代號"].nunique()
```

```python
group_by_stock.get_group("2330")["券商名稱"].unique()
```

# Regression

```python
tsmc = group_by_stock.get_group("2330")
```

```python
# 同一家券商可能在同一天發布對今年及去年的獲利預測，因此要去除重複的日期
tsmc = tsmc.groupby("券商名稱").apply(pd.DataFrame.drop_duplicates,subset=["日期"],keep="first")
```

```python
tsmc.pivot(index="日期",columns="券商名稱",values="目標價").fillna(method="ffill").plot()
```
